```python
import pandas as pd
import numpy as np

# week
cases_week = pd.read_csv('cases-week.csv')
cases_week = cases_week.drop(['Unnamed: 0'], axis = 1)
neighbor_week = pd.read_csv('neighbor-week.csv')
neighbor_week = neighbor_week.drop(['Unnamed: 0'], axis = 1)
state_week = pd.read_csv('state-week.csv')
state_week = state_week.drop(['Unnamed: 0'], axis = 1)

method_spot_week = {'weekid': [], 'method': [],'spot': [], 'districtid': []}
method_spot_week = pd.DataFrame(method_spot_week)

for i in range(len(cases_week)):
    if cases_week['cases'][i] > neighbor_week['neighbormean'][i]+neighbor_week['neighborstdev'][i]:
        temp = {'weekid': [cases_week['weekid'][i]], 'method': ['neighborhood'],'spot': ['hot'], 'districtid': [cases_week['districtid'][i]]}
        method_spot_week = method_spot_week.append(pd.DataFrame(temp))
    elif cases_week['cases'][i] < neighbor_week['neighbormean'][i]-neighbor_week['neighborstdev'][i]:
        temp = {'weekid': [cases_week['weekid'][i]], 'method': ['neighborhood'],'spot': ['cold'], 'districtid': [cases_week['districtid'][i]]}
        method_spot_week = method_spot_week.append(pd.DataFrame(temp))

for i in range(len(cases_week)):
    if cases_week['cases'][i] > state_week['statemean'][i]+state_week['statestdev'][i]:
        temp = {'weekid': [cases_week['weekid'][i]], 'method': ['state'],'spot': ['hot'], 'districtid': [cases_week['districtid'][i]]}
        method_spot_week = method_spot_week.append(pd.DataFrame(temp))
    elif cases_week['cases'][i] < state_week['statemean'][i]-state_week['statestdev'][i]:
        temp = {'weekid': [cases_week['weekid'][i]], 'method': ['state'],'spot': ['cold'], 'districtid': [cases_week['districtid'][i]]}
        method_spot_week = method_spot_week.append(pd.DataFrame(temp))

method_spot_week = method_spot_week.reset_index(drop=True)        
method_spot_week.to_csv('method-spot-week.csv')

# month
cases_month = pd.read_csv('cases-month.csv')
cases_month = cases_month.drop(['Unnamed: 0'], axis = 1)
neighbor_month = pd.read_csv('neighbor-month.csv')
neighbor_month = neighbor_month.drop(['Unnamed: 0'], axis = 1)
state_month = pd.read_csv('state-month.csv')
state_month = state_month.drop(['Unnamed: 0'], axis = 1)

method_spot_month = {'monthid': [], 'method': [],'spot': [], 'districtid': []}
method_spot_month = pd.DataFrame(method_spot_month)

for i in range(len(cases_month)):
    if cases_month['cases'][i] > neighbor_month['neighbormean'][i]+neighbor_month['neighborstdev'][i]:
        temp = {'monthid': [cases_month['monthid'][i]], 'method': ['neighborhood'],'spot': ['hot'], 'districtid': [cases_month['districtid'][i]]}
        method_spot_month = method_spot_month.append(pd.DataFrame(temp))
    elif cases_month['cases'][i] < neighbor_month['neighbormean'][i]-neighbor_month['neighborstdev'][i]:
        temp = {'monthid': [cases_month['monthid'][i]], 'method': ['neighborhood'],'spot': ['cold'], 'districtid': [cases_month['districtid'][i]]}
        method_spot_month = method_spot_month.append(pd.DataFrame(temp))

for i in range(len(cases_month)):
    if cases_month['cases'][i] > state_month['statemean'][i]+state_month['statestdev'][i]:
        temp = {'monthid': [cases_month['monthid'][i]], 'method': ['state'],'spot': ['hot'], 'districtid': [cases_month['districtid'][i]]}
        method_spot_month = method_spot_month.append(pd.DataFrame(temp))
    elif cases_month['cases'][i] < state_month['statemean'][i]-state_month['statestdev'][i]:
        temp = {'monthid': [cases_month['monthid'][i]], 'method': ['state'],'spot': ['cold'], 'districtid': [cases_month['districtid'][i]]}
        method_spot_month = method_spot_month.append(pd.DataFrame(temp))

method_spot_month = method_spot_month.reset_index(drop=True)        
method_spot_month.to_csv('method-spot-month.csv')

# overall
cases_overall = pd.read_csv('cases-overall.csv')
cases_overall = cases_overall.drop(['Unnamed: 0'], axis = 1)
neighbor_overall = pd.read_csv('neighbor-overall.csv')
neighbor_overall = neighbor_overall.drop(['Unnamed: 0'], axis = 1)
state_overall = pd.read_csv('state-overall.csv')
state_overall = state_overall.drop(['Unnamed: 0'], axis = 1)

method_spot_overall = {'overallid': [], 'method': [],'spot': [], 'districtid': []}
method_spot_overall = pd.DataFrame(method_spot_overall)

for i in range(len(cases_overall)):
    if cases_overall['cases'][i] > neighbor_overall['neighbormean'][i]+neighbor_overall['neighborstdev'][i]:
        temp = {'overallid': [cases_overall['overallid'][i]], 'method': ['neighborhood'],'spot': ['hot'], 'districtid': [cases_overall['districtid'][i]]}
        method_spot_overall = method_spot_overall.append(pd.DataFrame(temp))
    elif cases_overall['cases'][i] < neighbor_overall['neighbormean'][i]-neighbor_overall['neighborstdev'][i]:
        temp = {'overallid': [cases_overall['overallid'][i]], 'method': ['neighborhood'],'spot': ['cold'], 'districtid': [cases_overall['districtid'][i]]}
        method_spot_overall = method_spot_overall.append(pd.DataFrame(temp))

for i in range(len(cases_overall)):
    if cases_overall['cases'][i] > state_overall['statemean'][i]+state_overall['statestdev'][i]:
        temp = {'overallid': [cases_overall['overallid'][i]], 'method': ['state'],'spot': ['hot'], 'districtid': [cases_overall['districtid'][i]]}
        method_spot_overall = method_spot_overall.append(pd.DataFrame(temp))
    elif cases_overall['cases'][i] < state_overall['statemean'][i]-state_overall['statestdev'][i]:
        temp = {'overallid': [cases_overall['overallid'][i]], 'method': ['state'],'spot': ['cold'], 'districtid': [cases_overall['districtid'][i]]}
        method_spot_overall = method_spot_overall.append(pd.DataFrame(temp))

method_spot_overall = method_spot_overall.reset_index(drop=True)        
method_spot_overall.to_csv('method-spot-overall.csv')
```
