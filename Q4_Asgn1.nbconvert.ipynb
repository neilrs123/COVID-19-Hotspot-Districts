```python
import json 
import pandas as pd
import numpy as np
```

```python
# Opening JSON file 
f = open('neighbor-districts-modified.json') 
  
# returns JSON object as a dictionary 
data = json.load(f) 

# sorting districts alphabetically
districts_from_json = []
for i in data:
    districts_from_json.append(i)
districts_from_json = np.array(districts_from_json)
districts_from_json.sort()

# ids for districts
ids = np.arange(101,101+len(districts_from_json))

# Dictionary for storing ids of districts
Dict = {} 
for i in range(len(ids)):
    Dict[districts_from_json[i]]=ids[i]
```

```python
IDs_graph = {}
for dis in data:
    IDs_graph[Dict[dis]] = []
    for j in range(len(data[dis])):
        IDs_graph[Dict[dis]].append(Dict[data[dis][j]])
```

```python
cases_week = pd.read_csv('cases-week.csv')
cases_week = cases_week.drop(['Unnamed: 0'], axis = 1)

weeks = int(cases_week['weekid'].max())

Data1 = {'districtid': [],'weekid': [],'neighbormean': [], 'neighborstdev': [] }
neighbor_week = pd.DataFrame(Data1)

for i in range(1,weeks+1):
    start = len(IDs_graph)*(i-1)
    end = len(IDs_graph)*i
    numcases = {}
    for j in range(start,end):
        numcases[int(cases_week['districtid'][j])] = int(cases_week['cases'][j])
    neighbormean = []
    neighborstdev = []
    for j in range(101,101+len(IDs_graph)):
        neighborcases = []
        for k in range(len(IDs_graph[j])):
            neighborcases.append(numcases[IDs_graph[j][k]])
        neighborcases = np.array(neighborcases)
        if len(neighborcases)>0:
            neighbormean.append(np.mean(neighborcases))
            neighborstdev.append(np.std(neighborcases))
        else:
            neighbormean.append(0)
            neighborstdev.append(0)
    Data1 = {'districtid': np.arange(101,101+len(districts_from_json)),'weekid': i*np.ones(len(districts_from_json)),'neighbormean': neighbormean, 'neighborstdev': neighborstdev }
    Data1 = pd.DataFrame(Data1)
    neighbor_week = neighbor_week.append(Data1)

neighbor_week = neighbor_week.reset_index(drop=True)
neighbor_week.to_csv('neighbor-week.csv')
```
