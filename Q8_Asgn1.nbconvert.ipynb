```python
import json 
import pandas as pd
import numpy as np
```

```python
# Opening JSON file 
f = open('neighbor-districts-modified.json') 
  
# returns JSON object as a dictionary 
data = json.load(f) 

state_map = np.zeros(len(data))
```

```python
# week
zscore_week = pd.read_csv('zscore-week.csv')
zscore_week = zscore_week.drop(['Unnamed: 0'], axis = 1)

zscore_week = zscore_week.sort_values("neighborhoodzscore", axis = 0, ascending = False, kind='mergesort', inplace = False, na_position ='last')
zscore_week = zscore_week.sort_values("weekid", axis = 0, ascending = False, kind='mergesort', inplace = False, na_position ='last')
zscore_week = zscore_week.reset_index(drop=True)

top_week = {'weekid': [], 'method': [],'spot': [], 'districtid1': [], 'districtid2': [], 'districtid3': [], 'districtid4': [], 'districtid5': []}
top_week = pd.DataFrame(top_week)

weeks = int(zscore_week['weekid'].max())

for i in range(1,weeks+1):
    start = len(state_map)*(i-1)
    end = len(state_map)*i
    temp = {'weekid': [i], 'method': ['neighborhood'],'spot': ['hot'], 'districtid1': [zscore_week['districtid'][start]], 'districtid2': [zscore_week['districtid'][start+1]], 'districtid3': [zscore_week['districtid'][start+2]], 'districtid4': [zscore_week['districtid'][start+3]], 'districtid5': [zscore_week['districtid'][start+4]]}
    top_week = top_week.append(pd.DataFrame(temp))
    temp = {'weekid': [i], 'method': ['neighborhood'],'spot': ['cold'], 'districtid1': [zscore_week['districtid'][end-1]], 'districtid2': [zscore_week['districtid'][end-2]], 'districtid3': [zscore_week['districtid'][end-3]], 'districtid4': [zscore_week['districtid'][end-4]], 'districtid5': [zscore_week['districtid'][end-5]]}
    top_week = top_week.append(pd.DataFrame(temp))

zscore_week = zscore_week.sort_values("statezscore", axis = 0, ascending = False, kind='mergesort', inplace = False, na_position ='last')
zscore_week = zscore_week.sort_values("weekid", axis = 0, ascending = False, kind='mergesort', inplace = False, na_position ='last')
zscore_week = zscore_week.reset_index(drop=True)

for i in range(1,weeks+1):
    start = len(state_map)*(i-1)
    end = len(state_map)*i
    temp = {'weekid': [i], 'method': ['state'],'spot': ['hot'], 'districtid1': [zscore_week['districtid'][start]], 'districtid2': [zscore_week['districtid'][start+1]], 'districtid3': [zscore_week['districtid'][start+2]], 'districtid4': [zscore_week['districtid'][start+3]], 'districtid5': [zscore_week['districtid'][start+4]]}
    top_week = top_week.append(pd.DataFrame(temp))
    temp = {'weekid': [i], 'method': ['state'],'spot': ['cold'], 'districtid1': [zscore_week['districtid'][end-1]], 'districtid2': [zscore_week['districtid'][end-2]], 'districtid3': [zscore_week['districtid'][end-3]], 'districtid4': [zscore_week['districtid'][end-4]], 'districtid5': [zscore_week['districtid'][end-5]]}
    top_week = top_week.append(pd.DataFrame(temp))

top_week = top_week.reset_index(drop=True)        
top_week.to_csv('top-week.csv')
```
